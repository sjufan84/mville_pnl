```python
import pandas as pd
import numpy as np
import hvplot.pandas
import plotly_express as px
import re
```

```python
items_df = pd.read_csv('mville_inflation_new.csv',index_col = [0], parse_dates=True, infer_datetime_format=True)
cpi_df = pd.read_csv('CPIAUCSL.csv', index_col=[0], parse_dates=True, infer_datetime_format=True)
```

```python
# Drop the '$' and '%' from the values in the items_df dataframe except for in the column names
items_df = items_df.replace({'\$': '', '%': ''}, regex=True)
items_df
```

```python
# Convert the values in each dataframe to floats
items_df = items_df.astype(float)
cpi_df = cpi_df.astype(float)

# Sort the values in the items_df dataframe by the "Percent Increase" column
items_df = items_df.sort_values(by='Percent Increase', ascending=True)


# Generate a bar chart of the items_df dataframe that plots "Percent Increase" for each item using plotly express
items_df.hvplot.bar(x='PRODUCT NAME', y='Percent Increase', title='Percent Increase in Price of Items', xlabel='Item', ylabel='Percent Increase', rot=90, height=500, width=1000, color = 'Percent Increase', cmap='gist_heat_r')

```

```python
# Sort the dataframe by the "Price Change" column
items_df = items_df.sort_values(by='Price Increase', ascending=True)

# Generate a bar chart of the items_df dataframe that plots "Price Change" for each item
items_df.hvplot.bar(x='PRODUCT NAME', y='Price Increase', rot=90, height=500, width=1000, title='Margaritaville Food Cost Inflation from 2022 to 2023 in $', color = 'Price Increase', cmap='kr_r')
```

```python
# Change the "CPIAUCSL" column name to "CPI"
cpi_df = cpi_df.rename(columns={'CPIAUCSL': 'CPI'})
```

```python
# Use a line chart to plot the "CPIAUCSL" column from the cpi_df dataframe
cpi_df.hvplot.line(x='DATE', y='CPI', rot=90, height=500, width=1000, title='US Consumer Price Index (Cost of a basket of goods over the last year in the United States)', color = 'Orange')
```
