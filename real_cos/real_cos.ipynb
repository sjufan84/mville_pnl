```python
import pandas as pd
import numpy as np
```

```python
temp_df = pd.read_csv('./resources/temp_labor.csv')
```

```python
# Add a row at the top with the values of the column names
temp_df.loc[-1] = temp_df.columns
temp_df.index = temp_df.index + 1
temp_df = temp_df.sort_index()

```

```python
# Fill the nan values with 0
temp_df = temp_df.fillna(0)

```

```python
# Drop pthe last row
temp_df = temp_df.drop(temp_df.index[-1])

# Drop the 5th column and the "Unnamed: 8" column
temp_df = temp_df.drop(temp_df.columns[[4, 8]], axis=1)

temp_df.columns = ['Batch Number',	'Date', 'Posted',	'Type',	'Description',	'Stat',	'Amount',	'Vendor',	'Invoice Date']
```

```python
# Drop all rows where the "Stat" and "Description" columns are the same as another row
temp_df = temp_df.drop_duplicates(subset=['Stat', 'Description'], keep='first')

temp_df
```

```python
# Drop all of the ',' in the dataframe, and convert the "Description" column to a float
temp_df['Description'] = temp_df['Description'].str.replace(',', '')
temp_df['Description'] = temp_df['Description'].astype(float)
```

```python
temp_sum = temp_df['Description'].sum()

temp_sum
```

```python
food_df = pd.read_csv('./resources/food_cos.csv')
```

```python
food_df
```

```python
# fill the nan values with 0
food_df = food_df.fillna(0)

# Drop duplicate rows
food_df = food_df.drop_duplicates(keep='first')

food_df
```

```python
# Drop th "Unnamed: 8" column
food_df = food_df.drop(food_df.columns[[8]], axis=1)

food_df
```

```python
# Drop all the ',' in the "Amount" column
food_df['Amount'] = food_df['Amount'].str.replace(',', '')

# Convert the "Amount" column to a float
food_df['Amount'] = food_df['Amount'].astype(float)

# Sum the "Amount" column if the "Unnamed: 6" column is "DB"
debit_sum = food_df[food_df['Unnamed: 6'] == 'DB']['Amount'].sum()

# Sum the "Amount" column if the "Unnamed: 6" column is "CR"
credit_sum = food_df[food_df['Unnamed: 6'] == 'CR']['Amount'].sum()

# Subtract the credit sum from the debit sum
food_sum = debit_sum - credit_sum

food_sum

```
