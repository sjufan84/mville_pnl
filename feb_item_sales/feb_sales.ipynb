```python
import pandas as pd
import re
from datetime import date
import numpy as np
```

```python
sales_df = pd.read_csv('./resources/feb_ls_ltc_sales.csv')
```

```python
# Drop all rows with NaN values for the "Unnamed: 3" column
sales_df = sales_df.dropna(subset=['Unnamed: 3'])

sales_df

```

```python
# Fill all NaN values with 0
sales_df = sales_df.fillna(0)
```

```python
sales_df.columns = ['Check', 'Date','Time',	'Code',	'Name',	'Category',	'Department'	,'Course'	,'Price'	,'Qty',	'Item Total',	'Disc'	,'Outlet'	,'Order Employee',	'Disc. Employee'	,'Payment Method']

```

```python
# Drop all columns except "Date", "Code", "Name", "Qty", "Price", "Item Total", "Category", "Outlet"
sales_df = sales_df.drop(columns=['Check', 'Time',	'Department'	,'Course'	,'Disc'	,'Order Employee',	'Disc. Employee'	,'Payment Method'])

sales_df
```

```python
# Drop all rows where the "Code" column value does not contain "MAR"
sales_df = sales_df[sales_df['Code'].str.contains('MAR')]
```

```python
# Set the index to be the "Date" column and drop the "Date" column
sales_df = sales_df.set_index('Date')
```

```python
# Convert the index to a datetime object
sales_df.index = pd.to_datetime(sales_df.index)
```

```python
# The order of the columns should be "Name", "Code", "Price", "Qty", "Item Total", "Category", "OUtlet"
sales_df = sales_df[['Name', 'Code', 'Price', 'Qty', 'Item Total', 'Category', 'Outlet']]
sales_df
```

```python
sales_df.columns = ['Menu item name',	'Menu item code',	'Menu item list price'	,'Quantity sold'	,'Sales total excl. tax',	'Category', 'Outlet']
```

```python
# Create a new column called "Sales total incl. tax" and set it equal to the "Sales total excl. tax" column
sales_df['Sales total inc. tax'] = sales_df['Sales total excl. tax']

# Drop the "Sales total incl. tax" column and move the "Sales total inc. tax" column to the right of the "Sales total excl. tax" column
sales_df = sales_df[['Menu item name',	'Menu item code',	'Menu item list price'	,'Quantity sold'	,'Sales total excl. tax', 'Sales total inc. tax',	'Category', 'Outlet']]
sales_df

# 
```

```python
# Drop all ',' from the dataframes
sales_df = sales_df.replace(',','', regex=True)

# Convert the "Menu item list price" column to a float
sales_df['Menu item list price'] = sales_df['Menu item list price'].astype(float)

# Convert the "Sales total excl. tax" column to a float
sales_df['Sales total excl. tax'] = sales_df['Sales total excl. tax'].astype(float)

# Convert the "Sales total inc. tax" column to a float
sales_df['Sales total inc. tax'] = sales_df['Sales total inc. tax'].astype(float)

# Convert the "Quantity sold" column to an integer
sales_df['Quantity sold'] = sales_df['Quantity sold'].astype(float)

# Convert the columns names to string
sales_df.columns = sales_df.columns.astype(str)
```

```python
sales_df['Outlet'].unique()
```

```python
# Create a dictionary to hold the data we will concatenate with the dataframes
new_data = {
    'Location name': '',
    'Begin date': [date(2023, 2, 16)],
    'End date': [date(2023, 2, 16)],
    'Total revenue excl. tax': [30],
    'Total revenue inc. tax': [32]
}
```

```python
# Create a dictionary to store the data for each outlet
outlet_data = {}

# Loop through each outlet
for outlet_name in sales_df['Outlet'].unique():
    # Create a new dataframe for the current outlet
    outlet_df = sales_df[sales_df['Outlet'] == outlet_name].copy()
    
    # Add the outlet name to the dictionary
    outlet_data[outlet_name] = outlet_df
    
    # Print the dataframe for the current outlet
    print(outlet_name)
    print(outlet_df)


# Convert any tuple values in the "Outlet" column to a string
sales_df['Outlet'] = sales_df['Outlet'].apply(lambda x: x if type(x) != tuple else x[0])




```

```python
# For each of the outlet dataframes, populate the new data dictionary with the correct values and then concatenate the new data dictionary with the outlet dataframe
for outlet in outlet_data.values():
    new_data['Location name'] = str(outlet['Outlet'].unique()[0])
    new_data['Total revenue excl. tax'] = outlet['Sales total excl. tax'].sum()
    new_data['Total revenue inc. tax'] = outlet['Sales total inc. tax'].sum()

    # Sort the index by date
    outlet = outlet.sort_index()

    # Set the "Begin date" to the first date in the outlet dataframe
    new_data['Begin date'] = pd.to_datetime(outlet.index[0])

    # Set the "End date" to the last date in the outlet dataframe
    new_data['End date'] = pd.to_datetime(outlet.index[-1])

    new_data_df = pd.DataFrame.from_dict(new_data, orient='index')
    display(new_data_df)


    # Drop the "outlet" column
    outlet_df = outlet.drop(columns=['Outlet'])

    # Reset the index
    outlet_df = outlet_df.reset_index()

    # Drop the 'Date' column
    outlet_df = outlet_df.drop(columns=['Date'])
    outlet_df.shift(-2)

    # Concatenate the new data dictionary with the outlet dataframe with a space in between.  Shift the outlet dataframe to the left by two columns
    combined_df = pd.concat([new_data_df, pd.DataFrame(columns = new_data_df.columns), outlet_df])

    # Reset the index and move the column names to row 6
    combined_df = combined_df.reset_index()

    # Rename the columns to numbers starting with 0
    combined_df.columns = range(combined_df.shape[1])

    # Shift the dataframe beginning at row 6 down by 1 row and left by 2 columns
    combined_df.iloc[5:] = combined_df.iloc[5:].shift(2)

   
    

    
    # Insert the outlet column names into row 6
    combined_df.iloc[6, 2:len(outlet_df.columns)+2] = outlet_df.columns.astype(str)

    # Shift the top 4 rows to the right by 2 columns
    combined_df.iloc[:5] = combined_df.iloc[:5].shift(2, axis=1)

    # If a value in the dataframe is None, replace it with an empty string
    combined_df = combined_df.replace(np.nan, '', regex=True)

    # Drop the first 3 columns
    combined_df = combined_df.drop(columns=[0, 1])

    # Rename the columns to the numbers starting with 0
    combined_df.columns = range(combined_df.shape[1])



   

    

    # Export the dataframe to an excel file
    combined_df.to_excel(f'./venue_sales/{outlet["Outlet"].unique()[0]}.xlsx', index=False, header=False)

    

    





    



```
